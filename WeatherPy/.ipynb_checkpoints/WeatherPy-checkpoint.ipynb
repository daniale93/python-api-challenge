```python

import matplotlib.pyplot as plt
import requests
import pandas as pd
import numpy as np
from api_keys import weather_api_key
from citipy import citipy
import random
from scipy.stats import linregress

# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="


cities = []
# lat = []
# long = []
# temp = []

```

## Generate Cities List

```python

number_cities = 1500
random_array = np.arange(0, number_cities, 1)

lat = [random.uniform(-90, 90)   for value in random_array]
long = [random.uniform(-180, 180)   for value in random_array]

lat_long = zip(lat,long)

for lat_long in lat_long:
    city = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
    
    if city not in cities:
        cities.append(city)

print(f"There are {len(cities)} cities")
```

```python
city_two = []
cloudinesses = []
dates = []
humidities = []
lats = []
lngs = []
max_temps = []
wind_speeds = []
countries = []
count_one = 0
set_one = 1
for city in cities:
    try:
        response = requests.get(query_url + city.replace(" ","&")).json()
        cloudinesses.append(response['clouds']['all'])
        countries.append(response['sys']['country'])
        dates.append(response['dt'])
        humidities.append(response['main']['humidity'])
        lats.append(response['coord']['lat'])
        lngs.append(response['coord']['lon'])
        max_temps.append(response['main']['temp_max'])
        wind_speeds.append(response['wind']['speed'])
        if count_one > 48:
            count_one = 1
            set_one += 1
            city_two.append(city)
        else:
            count_one += 1
            city_two.append(city)
        print(f"Processing Record {count_one} of Set {set_one} | {city}")
    except Exception:
        print("City not found. Skipping...")
print("------------------------------\nData Retrieval Complete\n------------------------------")
```

```python

weather_dict = {
    "City":city_two,
    "Cloudiness":cloudinesses,
    "Country":countries,
    "Date":dates,
    "Humidity":humidities,
    "Lat":lats,
    "Lng":lngs,
    "Max Temp":max_temps,
    "Wind Speed":wind_speeds
}
```

```python
weather_dataframe = pd.DataFrame(weather_dict)
weather_dataframe.head


output_data_file = "Output/weather_df.csv"
weather_dataframe.to_csv(output_data_file)
```

## Linear Regression

```python
north_hemi_data = clean_weather[clean_weather['Lat'] >= 0] 

south_hemi_data = clean_weather[clean_weather['Lat'] < 0] 
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python

x_values = north_hemi_data['Lat']
y_values = north_hemi_data['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.title("Northern Hemisphere Latitude vs. Temperature Plot")
plt.xlabel("Latitude")
plt.ylabel("Temperature")
plt.savefig("Figures/fig5.png")
```
